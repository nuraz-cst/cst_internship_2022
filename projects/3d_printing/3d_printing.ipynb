# Project I - 3D Printing
## Nuraz Waris

For this project, I was tasked with creating a trough that would be used to effectively clean cell pallettes. In order to accomplish this, I used FreeCad, a 3D modeling software for macOS, to construct the model for the trough.

![FreeCad](https://upload.wikimedia.org/wikipedia/commons/f/f7/FreeCAD-logo.svg)

In order to install this software, I needed to download Homebrew first. Homebrew is a package manager that is compatible with macOS or linux.

![Homebrew](https://infosecmonkey.com/wp-content/uploads/2018/10/1538505807_featured.jpeg)

In order to install homebrew, I entered /bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)" into my device's terminal.

Next, I entered the following commands to download and install FreeCad on my device:
* brew install --cask freecad
* brew install freecad


This was my first experience utilizing FreeCad to create a 3D model. I quickly learned the basic functions of the software and got set to build this trough. 
1. I started off by using the sketch tool to create a sketch on the XZ plane - just a basic rectangle
2. I scaled the measurements of the perimeter of the trough using the vertical and horizontal dimensionalize tools - L: 14" W: 4"
3. I padded the sketch, essentially creating a rectangular prism - H: 6"
4. I created another sketch on the surface of the original prism - 0.5" away from the perimeter
5. I created a pocket within this sketch - P1: 1.5"
6. I repeated step four within this pocket, creating another sketch 0.5" away from the perimeter of the pocket
7. I created a final pocket within this sketch - P2: 4"

Below is an image of my original design:

Functionality wise, the design worked perfectly, but the client preferred the trough to be more narrow and slimmer. In order to accomplish this, I edited the dimensions of the width to be 2.5" instead of 4" leaving the dimensions as follows - 
* Sketch 1/Pad 1 = length: 14" , width: 2.5" , height: 6" 
* Sketch 2/Pocket 1 = length 13" , width: 2" , height 1.5"
* Sketch 3/Pocket 2 = length 12" , width: 1.5" , height 4"

In conclusion, this project was a complete success. The trough works effectively and is being used right this very minute. This project taught me a lot about 3D design and how to use FreeCad to create a 3D model. In addition, I learned how to download and install Homebrew and FreeCad on a device. 

There were some challenging aspects when it came to getting used to FreeCad's interface and what it had to offer. I wouldn't call myself an expert after one project, but over the course of creating this trough, I have an intermediate understanding of how to use FreeCad to design a 3D model and print it.
