* brew install pyenv | Installs pyenv (python environment)
* pyenv install 3.10.4 | Installs the 3.10.4 version of python
* pyenv global 3.10.4 | Configures the system to use python 3.10.4 globally
* python3.10 -m venv repo_name/projects/sanger_sequencing/venv | Creates a virtual python environment (venv) in a specific directory
* source repo_name/projects/sanger_sequencing/venv/bin/activate | Activates the virtual environment
* pip install jupyter lab | Installs jupyter lab into the virtual environment
* pip install biopython | Installs bipython into the virtual environment

The interface of jupyter lab is shown below:

As you can see in the image, there are a few options to choose from when using jupyter's lab. First, you can place your file in an existing folder or create a new folder in the left panel. In the right panel, we have a selection for a file type: notebook, console, or other (terminal, text file, markdown file, python file, and show contextual help)

For this project, I used the notebook option and selected Python 3 (ipykernel). Doing this will take you to the screen below:

In this interface, you use the cell type selection tool located at the top of the middle panel (where it says **code**). From here, you can switch between code, markdown, or raw for each cell type, depending on whichever one suits your needs more. For this project, I've mainly been using markdown because it's easy to use and quickly gets the job done.

I will be demonstrating some of the things you can do with markdown below. For code, it's simply python itself. 

If you'd like to put a heading for your notebook use # your_heading.

# Heading

For a subheading, use ## your_subheading

## Subheading

For text, just simply type as you normally would in a document. To **bold** or *italicize* use asterisks:
* Bold - ** bold ** (Without the spaces. If I didn't put a space, it would automatically bold it).
* Italicize - * italicize * (Same as before).

If you would like to start another paragraph, make sure there is a blank line in between the previous and following paragraph, otherwise it will conjoin.

As you saw previously, you can also make bulleted and numbered lists. For bulleted lists, use * word. For numbered lists, use 1. word , 2. word, etc.

Examples:

Grocery shopping list:
* Bread
* Eggs
* Milk
* Chicken

Top 5 cookies of all time:
1. Chocolate Chip Cookies
2. Sugar Cookies
3. Fudge Striped Cookies
4. Oreo Cookies
5. Peanut Butter Cookies


There are multiple ways of adding images into a jupyter notebook. These include:

1. Drag and drop files into a markdown cell
2. Copy/paste a URL of an image from the web
3. Use a local URL to an image that you will be keeping alongside your notebook, such as in the same git repo.
